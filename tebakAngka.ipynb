```python
#@title chipmode
import random

def tebak_angka():
    # Langkah 1: Input jumlah pemain dan nama pemain
    while True:
        try:
            jumlah_pemain = int(input("Masukkan jumlah pemain: "))
            if jumlah_pemain > 0:
                break
            else:
                print("Jumlah pemain harus lebih dari 0!")
        except ValueError:
            print("Masukan tidak valid! Masukkan angka.")

    pemain = []
    for i in range(jumlah_pemain):
        nama = input(f"Masukkan nama pemain ke-{i+1}: ")
        pemain.append({"nama": nama, "saldo": 100, "tebakan": None})  # Saldo awal 100

    # Langkah 2: Setiap pemain memasukkan taruhan
    for p in pemain:
        while True:
            try:
                taruhan = int(input(f"{p['nama']}, masukkan taruhan (saldo tersisa: {p['saldo']}): "))
                if taruhan <= p['saldo']:
                    p['taruhan'] = taruhan
                    break
                else:
                    print("Saldo tidak mencukupi!")
            except ValueError:
                print("Masukan tidak valid! Masukkan angka.")

    # Langkah 3: Angka acak
    angka_acak = random.randint(1, jumlah_pemain)

    # Langkah 4: Setiap pemain menebak angka
    angka_tertebak = set()  # Himpunan untuk melacak angka yang sudah ditebak
    for p in pemain:
        while True:
            sisa_angka = set(range(1, jumlah_pemain + 1)) - angka_tertebak
            try:
                tebakan = int(input(f"{p['nama']}, tebak angka ({sisa_angka}): "))
                if tebakan in sisa_angka:
                    p['tebakan'] = tebakan
                    angka_tertebak.add(tebakan)  # Menambahkan angka yang sudah ditebak ke himpunan
                    break
                else:
                    print("Angka tidak sesuai dengan jumlah pemain atau tidak valid! Coba lagi.")
            except ValueError:
                print("Masukan tidak valid! Masukkan angka.")

    # Langkah 5: Menampilkan hasil dan mengubah saldo
    pemenang = None
    for p in pemain:
        if p['tebakan'] == angka_acak:
            pemenang = p
            break

    if pemenang:
        print(f"Selamat! {pemenang['nama']} berhasil menebak angka {angka_acak} dan memenangkan taruhan.")
        for p in pemain:
            if p != pemenang:
                p['saldo'] -= p['taruhan']
                pemenang['saldo'] += p['taruhan']
                print(f"{p['nama']} kehilangan {p['taruhan']} ke {pemenang['nama']}. Saldo {p['nama']}: {p['saldo']}")
        print(f"Saldo {pemenang['nama']}: {pemenang['saldo']}")
    else:
        print(f"Tidak ada yang berhasil menebak angka {angka_acak}. Uang taruhan hangus.")

# Memulai permainan
tebak_angka()
```
